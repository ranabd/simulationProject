```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
import plotly.graph_objects as go
import plotly.offline as py

df = pd.read_csv('Covid19_Bangladesh.csv')

x=df["Date"]

y2 = df["Confirmed case"]
y3 = df["Death"]
y4 = df["Lab Test"]



fig = go.Figure()


fig.add_trace(go.Scatter(x=x,
                    y = y2,
                    mode = "lines+markers",
                    name = "Daily confirmed cases"
                    ))

fig.add_trace(go.Scatter(x=x,
                    y = y3,
                    mode = "lines+markers",
                    name = "Daily death cases"
                    ))

fig.add_trace(go.Scatter(x=x,
                    y = y4,
                    mode = "lines+markers",
                    name = "Daily lab tests"
                    ))



fig.update_layout(title = dict(text = "Line chart of confirmed, death and lab tests",
                              font_size = 20
                              ),
                 yaxis = dict(title = "Number of cases"),
                 xaxis = dict(title = "Date"))
```
