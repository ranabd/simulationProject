```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
import plotly.graph_objects as go
import plotly.offline as py

df = pd.read_csv('Covid19_Bangladesh.csv')

# number of daily death plotting
x=df[df["Death"] != 0]["Date"]
y = df[df["Death"] != 0]["Death"]


fig = go.Figure()

fig.add_trace(go.Bar(x=x,
                    y = y,
                    text = y,
                    textposition = "auto"))

fig.update_layout(title = dict(text = "Daily deaths",
                              font_size = 20
                              ),
                 yaxis = dict(title = "Number of deaths"),
                 xaxis = dict(title = "Date"))
```
